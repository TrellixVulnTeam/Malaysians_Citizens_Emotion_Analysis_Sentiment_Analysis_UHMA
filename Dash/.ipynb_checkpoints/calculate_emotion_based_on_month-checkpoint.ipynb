```python
#import library
import pandas as pd
```

```python
quater1_2020 = pd.read_csv('Jan-March-2020.csv')
quater2_2020 = pd.read_csv('April-Jun-2020.csv')
quater3_2020 = pd.read_csv('Jul-Sept-2020.csv')
quater4_2020 = pd.read_csv('Oct-Dec-2020.csv')
quater1_2021 = pd.read_csv('Jan-March-2021.csv')
quater2_2021 = pd.read_csv('Apr-Jun-2021.csv')
```

```python
quater1_2020.head(100)
```

```python
emotion_count = quater1_2020.groupby('emotions').count()
emotion_sum = sum(emotion_count.tweets)
emotion_count.tweets = (emotion_count.tweets * 100 / emotion_sum).round(2)
emotion_count.tweets
```

```python
df = pd.DataFrame(emotion_count.tweets)
```

```python
df.head(7)
```

```python
df.to_csv('quater1_2020.csv', sep=',', index=False)
```
