```python
import pandas as pd
import re
from html.parser import HTMLParser
```

```python
extracted_data = pd.read_csv('extracted_data.csv')
```

```python
extracted_data.head()
```

```python
def clean_tweets(tweet):
    #remove URL
    tweet = re.sub(r"http\S+", "", tweet)
    
    #remove HTML character
    tweet = HTMLParser().unescape(tweet)
    
    #remove username
    tweet = re.sub(f"@[^\s]+[\s]?",'',tweet)
    
    #remove special characters
    tweet = re.sub('[^ a-zA-Z0-9]', '',tweet)
    
    #remove Numbers
    tweet = re.sub('[0-9]', '', tweet)
    
    #remove newline 
    tweet = re.sub("\n", "", tweet)
    
    #Split attached words
    tweet = " ".join([s for s in re.split("([A-Z][a-z]+[^A-Z]*)",tweet) if s])
    
    #converting tweets to lower case
    tweet = tweet.lower()
    
    return tweet
```

```python
#apply function to tweets column 
extracted_data['tweets'] = extracted_data['tweets'].apply(clean_tweets)
```
