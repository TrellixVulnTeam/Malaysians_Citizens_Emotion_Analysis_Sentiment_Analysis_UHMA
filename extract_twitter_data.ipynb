```python
text_query = 'covid-19 OR COVID-19 OR KitaJagaKita OR stay safe OR stay home OR corona OR Coronavirus -filter:retweets'
coordinates = '3.140853,101.693207,400mi'
result_type = 'recent'
max_tweets = 1000

tweets = tw.Cursor(api.search, q=text_query, geocode=coordinates,
                  result_type = result_type, count=100).items(max_tweets)
```

```python
tweets_df = pd.DataFrame()

for tweet in tweets:
    try:
        text = api.get_status(id=tweet.id, tweet_mode='extended').full_text
    except:
        pass
    tweets_df = tweets_df.append(pd.DataFrame({'Datetime': [tweet.created_at], 
                                               'username': [tweet.user.screen_name], 
                                               'text': [text]}))
    
    tweets_df = tweets_df.reset_index(drop=True)
```
