```python
#import library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
```

```python
#load uncleaned train data
train_data = pd.read_csv('train_data.csv')
```

```python
train_data.head(175790)
```

```python
train_data.info()
```

```python
#cleaning data
def clean_tweets(tweet):
    #remove URL
    tweet = re.sub(r"http\S+", "", tweet)
    
    #remove username
    tweet = re.sub(f"@[^\s]+[\s]?",'',tweet)
    
    #remove special characters
    tweet = re.sub('[^ a-zA-Z0-9]', '',tweet)
    
    #remove Numbers
    tweet = re.sub('[0-9]', '', tweet)
    
    #converting tweets to lower case
    tweet = tweet.lower()
    
    return tweet
```

```python
#apply function to tweets column 
train_data['tweets'] = train_data['tweets'].apply(clean_tweets)
```

```python
train_data.head(175790)
```

```python
#store new cleaned training data into csv file
train_data.to_csv('cleaned_training_data.csv', sep=',', index=False)
```
