```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import pandas as pd
```

```python
angerData = pd.read_csv('anger.csv')
fearData = pd.read_csv('fear.csv')
happyData = pd.read_csv('happy.csv')
loveData = pd.read_csv('love.csv')
sadnessData = pd.read_csv('sadness.csv')
surpriseData = pd.read_csv('surprise.csv')
```

```python
plt.figure(figsize = (20,20))
wc = WordCloud(max_words = 1000 , width = 1600 , height = 800,
               collocations=False).generate(" ".join(angerData['tweets']))

plt.imshow(wc)
```
