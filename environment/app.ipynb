```python
import dash
import dash_core_components as dcc
import dash_html_components as html
import pandas as pd
import plotly.graph_objects as go

data = pd.read_csv("calculate_data.csv")

anger_percent = pd.read_csv("quater1_2020.csv")
fear_percent = pd.read_csv("quater2_2020.csv")
happy_percent = pd.read_csv("quater3_2020.csv")
love_percent = pd.read_csv("quater4_2020.csv")
sadness_percent = pd.read_csv("quater1_2021.csv")
surprise_percent = pd.read_csv("quater2_2021.csv")

#line chart
x = ["Jan-March 2020","Apr-Jun 2020","Jul-Sept 2020","Oct-Dec 2020","Jan-March 2021","Apr-Jun 2021"]
x1= ["Jan-March 2020","Apr-Jun 2020","Jul-Sept 2020","Oct-Dec 2020","Jan-March 2021","Apr-Jun 2021"]

y1 = anger_percent['percentage_anger']
y2 = fear_percent['percentage_fear']
y3 = happy_percent['percentage_happy']
y4 = love_percent['percentage_love']
y5 = sadness_percent['percentage_sadness']
y6 = surprise_percent['percentage_surprise']

figg = go.Figure()

figg.add_trace(go.Scatter(
    x=x, y=y1,
    line_color='rgb(0,100,80)',
    name='anger',
))

figg.add_trace(go.Scatter(
    x=x, y=y2,
    line_color='rgb(0,176,246)',
    name='fear',
))

figg.add_trace(go.Scatter(
    x=x, y=y3,
    line_color='rgb(231,107,243)',
    name='happy',
))

figg.add_trace(go.Scatter(
    x=x, y=y4,
    line_color='rgb(247,247,30)',
    name='love',
))

figg.add_trace(go.Scatter(
    x=x, y=y5,
    line_color='rgb(255,179,0)',
    name='sadness',
))

figg.add_trace(go.Scatter(
    x=x, y=y6,
    line_color='rgb(78,75,166)',
    name='surprise',
))

figg.update_traces(mode='lines')

figg.update_layout(title_text="Percentage of Malaysian's Citizen Emotion During COVID-19 based on group of Month from 2020 to 2021", 
                   xaxis_title='Group of Month', yaxis_title='Percentage(%)')


#stacked bar chart
figgg = go.Figure(go.Bar(x=x1,y=y1,name='anger'))
figgg.add_trace(go.Bar(x=x1,y=y2,name='fear'))
figgg.add_trace(go.Bar(x=x1,y=y3,name='happy'))
figgg.add_trace(go.Bar(x=x1,y=y4,name='love'))
figgg.add_trace(go.Bar(x=x1,y=y5,name='sadness'))
figgg.add_trace(go.Bar(x=x1,y=y6,name='surprise'))

figgg.update_layout(barmode='stack', title_text="Percentage of Malaysian's Citizens Emotion During COVID-19 based on group of Month from 2020 to 2021",
                   xaxis_title='Group of Month', yaxis_title='Percentage(%)')


#pie charts
labels = data['emotions']
values = data['percentage']

# Use `hole` to create a donut-like pie chart
fig = go.Figure(data=[go.Pie(labels=labels, values=values, hole=.3)])

fig.update_layout(title_text="Overall Percentage of Malaysian's Citizens Emotion During COVID-19 from 2020 to 2021")



external_stylesheets = [
    {
        "href": "https://fonts.googleapis.com/css2?"
        "family=Lato:wght@400;700&display=swap",
        "rel": "stylesheet",
    },
]

app = dash.Dash(__name__, external_stylesheets=external_stylesheets)
app.title = "Emotion Analysis"
app.layout = html.Div(
    children=[
        html.Div(
            children=[
                html.P(children="üò°üòÉüò∞ü•∞‚òπÔ∏èüò≤", className="header-emoji"),
                html.H1(
                    children="Malaysian's Citizen Emotion Analysis During COVID-19 Outbreak", className="header-title"
                ),
                html.P(
                    children="Analyze the emotion of Malaysians"
                    " by using sentiment analysis"
                    " during COVID-19 Outbreak in Twitter",
                    className="header-description",
                ),
            ],
            className="header",
        ),
        html.Div(
            children=[
                html.Div(
                    children=dcc.Graph(config={"displayModeBar": False}, 
                                       figure=figg),
                    className="card",
                ),
                html.Div(
                    children=dcc.Graph(config={"displayModeBar": False}, 
                                       figure=figgg),
                    className="card",
                ),
                html.Div(
                    children=dcc.Graph(config={"displayModeBar": False}, 
                                       figure=fig),
                    className="card",
                ),
            ],
            className="wrapper",
        ),
    ]
)

if __name__ == "__main__":
    app.run_server(debug=False,
                   host = '127.0.0.1')
```
